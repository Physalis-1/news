### Simple generation

```python
def filter_resuls(nr):
    return [x[:x.find("<|endoftext|>")] for x in nr]
```

Greedy decoding

```python
#@title Текст заголовка по умолчанию
import pandas as pd
datafram=pd.DataFrame()
import copy
texts=["Новость: специалисты по информационной безопасности заявили",
"Новость: в ФСБ опровергли сообщения",
"Новость: взломаны сервера одной из крупнейших компаний",
"Новость: ФСТЭК опубликовал",
"Новость: Российсике хакеры снова",
"Новость: ФСТЭК вводит новые"]
for j in range (0,len(texts)):
  for i in range (0, 10):
    element=filter_resuls(gpt.generate(
    text=texts[j],
    top_k=5,
    top_p=0.95,
    temperature=1.2,
    num_return_sequences=1,
    do_sample=True,
    max_length=400,
    min_length=200,
    no_repeat_ngram_size=3,
    repetition_penalty=2.,
    ))
    print('step=',j,' нагенерено',i+1)
# print(element)
    datafram=pd.concat([datafram,pd.DataFrame({'News':[copy.deepcopy(element)], 'InfSec':[1],'FAKE?':[1],'Link':[copy.deepcopy('https://fake_news.ru')]})])
```

```python
datafram.to_csv('dataset7.csv', encoding='utf-8', index=False)
```
